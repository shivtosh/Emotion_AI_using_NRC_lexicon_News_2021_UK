```python
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words = 'english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

 Word clouds for positive , negative and neutral texts (according to text blob sentiment analysis)

```python
def corpus_by_polarity(df):
    pos_text=""
    neg_text=""
    neu_text=""
    for index in range(len(df)):
        if (df.loc[index,'polarity']>0):
            df.loc[index,'pol_tb']="positive"
            pos_text=pos_text+df.loc[index,'text']
        elif (df.loc[index,'polarity']<0):
            df.loc[index,'pol_tb']="negative"
            neg_text=neg_text+df.loc[index,'text']
        elif (df.loc[index,'polarity']==0):
            df.loc[index,'pol_tb']="neutral"
            neu_text=neu_text+df.loc[index,'text']
    return (pos_text,neg_text,neu_text)
        
    
```

```python
pos_text,neg_text,neu_text=corpus_by_polarity(df)
```
